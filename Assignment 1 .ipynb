# 11) Remove outliers and keep outliers (does it have an effect on the final predictive model)?

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# Assuming df is your DataFrame
X = df.drop('2022 population', axis=1)
y = df['2022 population']

# Encode target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

def remove_outliers(df):
    numeric_cols = df.select_dtypes(include=['number']).columns
    Q1 = df[numeric_cols].quantile(0.25)
    Q3 = df[numeric_cols].quantile(0.75)
    IQR = Q3 - Q1
    return df[~((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

# Check if there are any non-numeric columns in X
non_numeric_cols = X.select_dtypes(include=['object']).columns
if not non_numeric_cols.empty:
    # If there are categorical variables, encode them
    X_encoded = pd.get_dummies(X, columns=non_numeric_cols, drop_first=True)
else:
    X_encoded = X

# Split the data into training and testing sets with the original data
X_train_orig, X_test_orig, y_train_orig, y_test_orig = train_test_split(X_encoded, y_encoded, test_size=0.3, random_state=42)

# Build and evaluate the model with the original data
model_orig = RandomForestClassifier(random_state=42)
model_orig.fit(X_train_orig, y_train_orig)
predictions_orig = model_orig.predict(X_test_orig)
accuracy_orig = accuracy_score(y_test_orig, predictions_orig)
print("Accuracy with original data:", accuracy_orig)

# Remove outliers and evaluate the model with outlier-removed data
X_no_outliers = remove_outliers(X_encoded)
X_no_outliers_reset = X_no_outliers.reset_index(drop=True)  # Reset index
y_no_outliers = y_encoded[X_no_outliers_reset.index]

# Split the outlier-removed data into training and testing sets
X_train_no_outliers, X_test_no_outliers, y_train_no_outliers, y_test_no_outliers = train_test_split(
    X_no_outliers_reset, y_no_outliers, test_size=0.3, random_state=42
)

# Build and evaluate the model with outlier-removed data
model_no_outliers = RandomForestClassifier(random_state=42)
model_no_outliers.fit(X_train_no_outliers, y_train_no_outliers)
predictions_no_outliers = model_no_outliers.predict(X_test_no_outliers)
accuracy_no_outliers = accuracy_score(y_test_no_outliers, predictions_no_outliers)
print("Accuracy without outliers:", accuracy_no_outliers)
```

# 12) Remove 1%, 5%, and 10% of your data randomly and impute the values back using at least 3 imputation methods. How well did the methods recover the missing values?  That is to remove some data, check the % error on residuals for numeric data and check for bias and variance of the error.

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

df_numeric = df.select_dtypes(include=[np.number])

def remove_data_randomly(df, percent):
    df_removed = df.copy()
    for column in df_removed.columns:
        num_missing = int(len(df_removed) * percent)
        missing_indices = np.random.choice(df_removed.index, num_missing, replace=False)
        df_removed.loc[missing_indices, column] = np.nan
    return df_removed

def impute_and_evaluate(df_original, df_removed, strategy):
    imputer = SimpleImputer(strategy=strategy)
    df_imputed = pd.DataFrame(imputer.fit_transform(df_removed), columns=df_removed.columns)
    mse = mean_squared_error(df_original.values.flatten(), df_imputed.values.flatten())
    return mse

percentages = [0.01, 0.05, 0.10]
strategies = ['mean', 'median', 'most_frequent']

for percent in percentages:
    df_removed = remove_data_randomly(df_numeric, percent)
    print(f"\n--- Removing {percent*100}% of data ---")
    for strategy in strategies:
        mse = impute_and_evaluate(df_numeric, df_removed, strategy)
        print(f"Strategy: {strategy}, MSE: {mse}")
```
